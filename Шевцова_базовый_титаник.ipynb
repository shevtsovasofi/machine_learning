Принцип работы one-hot-encoding:

- создаем матрицу из двух векторов размерностью 10

**ОТВЕТ**

для работы с этим методом доступны:
- mean: замена пропущенных значений, используя среднее значение по каждому столбцу. Может использоваться только с *числовыми данными*.
- median: замена пропущенных значений, используя медиану по каждому столбцу. Может использоваться только с *числовыми данными*.
- most_frequent: замена пропущенных значений, используя наиболее частое значение по каждому столбцу. Может использоваться со *строками или числовыми данными*. Если таких значений несколько, возвращается только наименьшее.
- constant: замена пропущенных значений на fill_value. Может использоваться со *строками или числовыми данными*.

Значит, для работы с категориальными данными подойдут только most_frequent и constant

**Feature Engineering**

> Данные - это пища для вашей модели

Слишком шумные данные вызовут отравление модели (кстати, один из методов киберпреступности, построенный на работе с данными, так и называется - data poisoning!)

Недостаточность в данных может стать причиной недообучения (underfitting), поэтому мы можем не только дропать, но и добавлять фичи, т.е. признаки, на которых будет учиться наша модель

Этот процесс называется **feature engineering**

````python
# Добавим фичу Family size = SibSp + Parch + 1 (единица обозначает самого пассажира)
new_df['FamilySize'] = new_df['SibSp'] + new_df['Parch'] + 1

# Фича Alone для тех, кто путешествует один
new_df['IsAlone'] = (new_df['FamilySize'] == 1).astype(int)

new_df.head()
````

Добавьте любую новую фичу в датафрейм, которая легко выводится из данных

Например, можно вытащить обращения к пассажирам из столбца `Name` из исходной таблицы: `Mr., Mrs., Dr.`

Для этого изучите внимательно данные, познакомьтесь с описанием датасета на Kaggle

- https://www.kaggle.com/c/titanic/

Вам нужно самостоятельно решить, что можно извлечь из этих данных, как назвать новую фичу и избавиться от категориальных меток

````python
print(new_df.columns)
````

````python
##вытаскиваем обращения
import re
def extract_status(name):
    match = re.search(r',\s([A-Za-z]+)\.', name) #ищем статусы в имени при помощи regex
    if match:
        return match.group(1)
    return np.nan

new_df['Status'] = df['Name'].apply(extract_status) # применяем функцию для извлечения статуса и добавляем результат в новый датаферйем

imputer = SimpleImputer(strategy='most_frequent')


new_df['Status'] = new_df['Status'].apply( # заменяем наны в статусах на самые частые значения
    lambda status: status if pd.notna(status) else imputer.fit_transform(new_df[['Status']]).flatten()[0]
)

print(set(new_df['Status']))  # выводим список статусов (уникальных)

new_df.head(10)
#итог: получились вот такие статусы: {'Rev', 'Don', 'Mr', 'Mme', 'Lady', 'Sir', 'Mlle', 'Mrs', 'Col', 'Master', 'Major', 'Capt', 'Jonkheer', 'Ms', 'Dr', 'Miss'}
#- это многовато, поэтому мы их разделим только на 2 группы - титулованные и простые люди
````

````python
#пусть титулованными будут: rev, don, col, major, capt, jonkheer, dr, sir
#тогда остальные - простые люди
high_status_titles = {'Rev', 'Don', 'Col', 'Major', 'Capt', 'Jonkheer', 'Dr', 'Sir'}

# создаём новый столбец, где титулованным  присваиваем High, остальным Low
new_df['HighStatus_Label'] = new_df['Status'].apply(lambda x: 'High' if x in high_status_titles else 'Low')

label_encoder = LabelEncoder()
new_df['Is_HighStatus'] = label_encoder.fit_transform(new_df['HighStatus_Label'])

if label_encoder.classes_[0] == 'High':
    new_df['Is_HighStatus'] = 1 - new_df['Is_HighStatus']

# Проверяем результат, высокопоставленные выделяются
new_df[['Status', 'HighStatus_Label', 'Is_HighStatus']].head(50)
````

````python
new_df.drop(columns=['Status', 'HighStatus_Label'], inplace=True)

new_df.head(50)
 #на всякий случай проверила итоговую табличку new.df
````

#### Шаг 4: обучение

Создаем пары `data - label`

````python
X = new_df.drop(columns=['Survived']) # данные = все столбцы кроме метки Survived
y = new_df['Survived'] # метки = только столбец Survived
````

Вспомните прошлое занятие: примените метод `train_test_split`

Создайте 2 выборки данных в соотношении 80 на 20

````python
X_train, X_test, y_train, y_test = (train_test_split(X, y, test_size=0.2, train_size=0.8))
````

Изучите информацию о том, как работает логистическая регрессия и ответьте на вопрос: какую роль играет `StandardScaler` в работе с логистической регрессией?

````python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
````

Обучение, как и в случае работы с линейной регрессией, займет всего пару строк

Опишите основной принцип работы логистической регрессии своими словами (2-3 предложения достаточно)

Предварительно изучите материал: https://web.stanford.edu/~jurafsky/slp3/5.pdf

````python
log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)

# предсказания
y_pred_log = log_reg.predict(X_test_scaled)

# вывод репорта
print("Logistic Regression Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_log))
print(classification_report(y_test, y_pred_log))
````

Повторим магию и обучим наивные байесовский классификатор. Что значит `Gaussian` в названии метода `GaussianNB`? Изучите документацию метода, чтобы ответить на этот вопрос

Почему для этого метода не нужен StandardScaler?

````python
nb = GaussianNB()
nb.fit(X_train, y_train)

y_pred_nb = nb.predict(X_test)

print("Naïve Bayes Performance:")
print("Accuracy:", accuracy_score(y_test, y_pred_nb))
print(classification_report(y_test, y_pred_nb))
````

Мы молодцы! Отрисуем результат с помощью матрицы ошибок

````python
conf_matrix_log = confusion_matrix(y_test, y_pred_log)
conf_matrix_nb =  confusion_matrix(y_test, y_pred_nb)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(conf_matrix_log, annot=True, fmt="d", cmap="Blues", ax=axes[0])
axes[0].set_title("Logistic Regression Confusion Matrix")

sns.heatmap(conf_matrix_nb, annot=True, fmt="d", cmap="Reds", ax=axes[1])
axes[1].set_title("Naïve Bayes Confusion Matrix")

plt.show()
````
